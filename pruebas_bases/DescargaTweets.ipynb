```python
def sin_https(lista):
    lista_def=lista
    for i in range(0,len(lista)):
        try:
            pal=lista[i].strip()
            word='https:'
            if word in pal:
                s=pal.find(word)
                e=s+len(word)
                while pal[e]!=" ":
                    if e<(len(pal)-1):
                        e=e+1
                    else:
                        break
                if s!=0 and e!=len(pal):
                    pal_final=pal[0:(s-1)]+pal[e:len(pal)]
                elif s!=0 and e==len(pal):
                    pal_final=pal[0:(s-1)]
                else:
                    pal_final=pal[e:len(pal)]
                lista_def[i]=pal_final
        except:
            print("revisar",i)
    return(lista_def)
```

## Embedding de cada frase de Twitter

```python
import gensim
```

```python
import numpy as np
```

```python
wv_embeddings = gensim.models.KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True, limit=500000)
```

```python
def tweet_to_vec(question, embeddings, dim=300):
    """
        question: string
        embeddings: llave=palabra, valor=vectorizacion
        dim: size de la representacion

        resultado= pregunta vectorizada en dicha dimension 
    """
    result = np.zeros(dim)
    cnt = 0
    # pregunta se convierte en lista
    words = question.split()
    for word in words:
        # Buscamos cada palabra de la lista en el diccionario
        if word in embeddings:
            result += np.array(embeddings[word])
            cnt += 1
    if cnt != 0:
        result /= cnt
    return result
```

```python
tweet_vec= []
for tweet in lista_v6:
    # Strip elimina el primer espacio en el string
    question = tweet.strip()
    # Cada pregunta la pasamos al testeo
    answer = tweet_to_vec(question, wv_embeddings)
    # Imprimimos cada pregunta con su vectorizacion 
    tweet_vec.append(answer)
    #print(question2vec_result)
```

Medida coseno de cada arreglo con los demas

```python
def medida_coseno(lista):
    lista_total=[]
    for i in lista:
        lista_coseno=[]
        for j in lista:
            if (i-j).all():
                a=cosine_similarity([i],[j])
                lista_coseno.append(a)
        lista_total.append(lista_coseno)
    return(lista_total)
```
